```python
def RectSine(amplitude, period, duration, phase, offset, delay):
    DATA_LENGTH = 32768
    ClockPeriod = 2e-9
    Points = np.linspace(-delay, ClockPeriod*(DATA_LENGTH-1)-delay, 32768)
    #print Points
    Pulse = np.where(Points>=0, np.where(Points<duration, (amplitude*np.sin(2*np.pi*Points/period+phase)).astype(int).tolist(), 0), 0) + int(offset)
    #print Pulse
    return Pulse

def DoubleRectSine(amplitude, period, duration, phase, offset, delay1, delay2):
    DATA_LENGTH = 32768
    ClockPeriod = 2e-9
    Points = np.linspace(0, ClockPeriod*(DATA_LENGTH-1), 32768)
    #print Points
    Pulse = np.where(Points>=delay1, np.where(Points<duration+delay1, (amplitude*np.sin(2*np.pi*Points/period+phase)).astype(int).tolist(), 0), 
                     0)+np.where(Points>=delay2, np.where(Points<duration+delay2, (amplitude*np.sin(2*np.pi*Points/period+phase)).astype(int).tolist(), 0), 0)+int(offset)
    #print Pulse
    return Pulse
def GaussSine(amplitude, period, length, phase, offset, delay):
    DATA_LENGTH = 32768
    ClockPeriod = 2e-9
    WindowSizeSigma = 6
    #Sigma = fwhm/np.sqrt(8*np.log(2))
    #WindowWidth = WindowSizeSigma*Sigma
    WindowWidth = length
    Sigma = WindowWidth/WindowSizeSigma
    Points = np.linspace(0, ClockPeriod*(DATA_LENGTH-1), 32768)
    envelope = amplitude*np.exp(-0.5*(Points-delay-WindowWidth/2)**2/Sigma**2)
    Pulse = (envelope*np.sin(2*np.pi*Points/period+phase)+offset).astype(int).tolist()
    return Pulse

def DoubleGaussSine(amplitude, period, length, phase, offset, delay1, delay2):
    DATA_LENGTH = 32768
    ClockPeriod = 2e-9
    WindowSizeSigma = 6
    #Sigma = fwhm/np.sqrt(8*np.log(2))
    #WindowWidth = WindowSizeSigma*Sigma
    WindowWidth = length
    Sigma = WindowWidth/WindowSizeSigma
    Points = np.linspace(0, ClockPeriod*(DATA_LENGTH-1), 32768)
    envelope = amplitude*np.exp(-0.5*(Points-delay1-WindowWidth/2)**2/(Sigma*0.5)**2)+amplitude*np.exp(-0.5*(Points-delay2-WindowWidth/2)**2/(Sigma*0.5)**2)
    Pulse = (envelope*np.sin(2*np.pi*Points/period+phase)+offset).astype(int).tolist()
    return Pulse

def RectSine2(amplitude, period, duration, phase, offset_off, offset_on, delay):
    DATA_LENGTH = 32768
    ClockPeriod = 2e-9
    Points = np.linspace(-delay, ClockPeriod*(DATA_LENGTH-1)-delay, 32768)
    #print Points
    Pulse = np.where(Points>=0, np.where(Points<duration, (amplitude*np.sin(2*np.pi*Points/period+phase)+(offset_on-offset_off)).astype(int).tolist(), 0), 0) + int(offset_off)
    #print Pulse
    return Pulse
```

```python
hardware_samplerate = 500e6
nums = 2000
ro_pulse_length = 0.05e-6
ro_readout_length = 1e-6
hardware_segsize = int(2**np.ceil(np.log(ro_readout_length*hardware_samplerate)/np.log(2))*2)
posttrigger = 3*hardware_segsize/4
amp = 2000 
offs = 50
timeout = 10000
reprate = 1/PulsePeriod
software_averages = 100
#software_averages = 500
software_binning = 8
binned_segsize = hardware_segsize/software_binning
center_freq_point = int(settings_ro[3]['F']/hardware_samplerate*hardware_segsize)
compute_covariations = False

pulse_lengths = np.linspace(0, 0.4e-6, 41)
max_pulse_length = np.max(pulse_lengths)
excite_powers = np.linspace(0.1, 0.5, 5)
excite_amplitudes = np.sqrt(excite_powers)

lo1.set_status(1)
pna.set_cw_freq(ro_freq-settings_ro[3]['F'])
sample = np.zeros((binned_segsize, len(pulse_lengths), len(excite_powers)))
if compute_covariations:
    covariation_matrix = np.zeros((binned_segsize, binned_segsize, len(pulse_lengths), len(excite_powers)))
S21 = np.zeros((len(pulse_lengths), len(excite_powers)), dtype=np.complex)
lo1.set_frequency(excite_freq-settings_excite[1]['F'])
#ro_results_output = open('C:/qtlab/ReiData/data/Feb 19 2016/rabi_dump/dump1.pkl', 'w')

try:
    for amplitude_id, amplitude in enumerate(excite_amplitudes):
        for pulse_length_id, excite_pulse_length in enumerate(pulse_lengths):
            excitePulseI = RectSine(settings_excite[1]['A']*amplitude, 1./settings_excite[1]['F'], excite_pulse_length, np.pi/2, 
                                settings_excite[1]['O'], max_pulse_length-excite_pulse_length)
            excitePulseQ = RectSine(settings_excite[2]['A']*amplitude, 1./settings_excite[2]['F'], excite_pulse_length, 
                                settings_excite[2]['P']+np.pi/2, settings_excite[2]['O'], max_pulse_length-excite_pulse_length)
            ROPulseI = RectSine(settings_ro[3]['A'], 1./settings_ro[3]['F'], ro_pulse_length, np.pi/2, 
                                settings_ro[3]['O'], max_pulse_length)
            ROPulseQ = RectSine(settings_ro[4]['A'], 1./settings_ro[4]['F'], ro_pulse_length, 
                                settings_ro[4]['P']+np.pi/2, settings_ro[4]['O'], max_pulse_length)
            # setting trigger to start readout after excitation pulses
            triggerPulse = [0, zero_delay+int(max_pulse_length/1e-9), 3, 100, 3, 1, 0, 4]
            SyncPulseProg = awg.intToBytes(triggerPulse)
            awg.ProgPulse(3, SyncPulseProg)
            awg.LoadRAM(1, excitePulseI)
            awg.LoadRAM(2, excitePulseQ)
            awg.LoadRAM(3, ROPulseI)
            awg.LoadRAM(4, ROPulseQ)
            for software_average_id in range(software_averages):
                daq0.stop()
                daq0.set_timeout(timeout)
                daq0.set_spc_samplerate(hardware_samplerate)
                daq0.set_ext_clock(500e6)
                daq0.set_trigger_ext0_level0(1000) # set trigger zero value to 200 mV
                daq0.set_trigger_ext0_level1(2500) # set trigger unity value to 600 mV
                daq0.trigger_termination_50Ohm()
                daq0.init_channel0_multiple_recording(nums=nums, segsize=hardware_segsize, posttrigger=posttrigger, amp=amp, offs=offs)
                daq0.buffer_setup()
                acq_start = time.time()
                daq0.start_with_trigger_and_waitready()
                ro_start= time.time()
                data = np.sum(np.reshape(daq0.readout_singlechannel_multimode_bin(), (nums, binned_segsize, software_binning)), axis=2).T
                fft_start = time.time()
                data_fft = np.fft.fft(data, axis=0)
                mean_start = time.time()
                sample[:,pulse_length_id, amplitude_id] += np.sum(data, axis=1)
                covariation_start = time.time()
                if compute_covariations:
                    covariation_matrix[:,:,pulse_length_id, amplitude_id] += np.einsum('ij,kj->ik', data, np.conj(data))
                #pickle.dump(data_fft[center_freq_point-8:center_freq_point+8, :], ro_results_output)
                cycle_end = time.time()
                S21[pulse_length_id, amplitude_id] += + np.sum(data_fft, axis=1)[center_freq_point]
        #plt.cla()
        #plt.plot(sample[:, pulse_length_id])
        #plt.pause(0.1)
        #print 'acq time: {0:5.3g}, ro time: {1:5.3g}, fft time: {2:5.3g}, mean time: {3:5.3g}, covariation time{4:5.3g}'.format(
        #    ro_start-acq_start, fft_start-ro_start, mean_start - fft_start, covariation_start - mean_start, cycle_end-covariation_start)
                daq0.stop()
                plt.clf()
                plt.pcolor(excite_amplitudes[:amplitude_id+1], pulse_lengths, np.abs(S21[:,:amplitude_id+1]))
                plt.colorbar()
                plt.pause(0.1)
            last_pulse_length_point = pulse_length_id
        last_amplitude_point = amplitude_id
#ro_results_output.close()
finally:
    #excite_freq = excite_freq_old
    print('finished')
```
