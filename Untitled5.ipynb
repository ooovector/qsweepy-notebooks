```python
import numpy as np
%pylab inline
```

```python
def sample_frequency(phi_min, phi_max, num_points):
    phi = np.linspace(phi_min, phi_max, num_points)
    f = np.sqrt((1+cos(phi))/(1-cos(phi)))*np.sign(np.floor(phi/np.pi)%2-0.5)
    return f
```

```python
f = sample_frequency(np.pi*0.125, np.pi*1.875, 25)
```
