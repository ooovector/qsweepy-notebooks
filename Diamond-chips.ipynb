```python
S21 = save_exdir.load_exdir(db.Data[1900].filename)
```

```python
data = S21.datasets['S-parameter'].data
delay = 63.7e-9
frequencies = S21.datasets['S-parameter'].parameters[1].values
data_nodelay = data*np.exp(1j*2*np.pi*delay*frequencies)
```

```python
plt.figure()
plt.pcolormesh(frequencies*1e-9, 
               S21.datasets['S-parameter'].parameters[0].values*1e6, 
               #np.unwrap(np.angle(data*np.exp(+1j*2*np.pi*delay*frequencies))))
               np.angle(data_nodelay), cmap='jet')
plt.xlabel('Frequency (GHz)')
plt.ylabel('DC bias (μA)')
plt.title('∠$S_{21}$ (rad)')
plt.colorbar()
```
