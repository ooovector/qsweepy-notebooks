```python
phases = data[1]['S-parameter'][1][0]
S21 = data[1]['S-parameter'][2]
hists = []
for phase_id, phase in enumerate(phases):
    _S21 = S21[phase_id,:]
    _S21_r = np.real(_S21)
    _S21_i = np.imag(_S21)
    hists.append(np.histogram2d(_S21_r, _S21_i, bins=100))
```

```python
hists[0][0]
```

```python
plt.pcolormesh(hists[0][1], hists[1][2], hists[0][0])
plt.title('Phase: {}'.format(phases[0]))
plt.show()
```
